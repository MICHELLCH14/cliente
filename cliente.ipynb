```python
class Cliente:
    def __init__(self, nombre, apellido, cedula, email, vip, direccion, telefono, celular, edad):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.email = email
        self.vip = vip
        self.direccion = direccion
        self.telefono = telefono
        self.celular = celular
        self.edad = edad

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor

    @property
    def apellido(self):
        return self._apellido

    @apellido.setter
    def apellido(self, valor):
        self._apellido = valor

    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, valor):
        if not isinstance(valor, str) or not valor.isdigit() or len(valor) != 10:
            raise ValueError("La cedula debe contener exactamente 10 digitos numericos")
        self._cedula = valor

    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, valor):
        self._email = valor

    @property
    def vip(self):
        return self._vip

    @vip.setter
    def vip(self, valor):
        self._vip = valor

    @property
    def direccion(self):
        return self._direccion

    @direccion.setter
    def direccion(self, valor):
        self._direccion = valor

    @property
    def telefono(self):
        return self._telefono

    @telefono.setter
    def telefono(self, valor):
        self._telefono = valor

    @property
    def celular(self):
        return self._celular

    @celular.setter
    def celular(self, valor):
        self._celular = valor

    @property
    def edad(self):
        return self._edad

    @edad.setter
    def edad(self, valor):
        if not isinstance(valor, int) or valor < 18:
            raise ValueError("La edad debe ser un numero entero mayor a 18")
        self._edad = valor

    def __str__(self):
        return f"Cliente: {self.nombre} {self.apellido} | Cedula: {self.cedula} | Edad: {self.edad} | VIP: {self.vip}"


if __name__ == "__main__":
    mi_cliente = Cliente("Juan", "Perez", "0987654321", "juan@email.com", True, "Calle Falsa 123", "042123456", "0991234567", 25)
    print(mi_cliente)
```
